## Working with pandas & DataFrames
One of the best options for working with tabular data in Python is to use the Python library `pandas`. The `pandas` library provides data structures, produces high quality plots with `matplotlib` and integrates nicely with other libraries that use NumPy (which is another Python library) arrays.

**Python doesnâ€™t load all of the libraries available to it by default.** We have to add an import statement to our code in order to use library functions. To import a library, we use the syntax `import libraryName`. 

If we want to give the library a nickname to shorten the command, we can add `as nickNameHere`. An example of importing the pandas library using the common nickname `pd` is below.

````
import pandas as pd
````

### Use `pandas` to read CSV-formatted data.

In Part 1, we loaded the gapminder data into Python's memory space by using the built-in `open` and `read` functions. This was less than efficient, since `read` loads the contents of the file as a single string, leaving it up to the programmer to figure out how to parse it.

Python has a `csv` library that can be used to parse CSV files. But we won't use that here, because the `pandas.read_csv()` function provides a further level of abstraction, allowing us seamlessly to convert our CSV data into a complex Python object called a `DataFrame`.


````
pd.read_csv('data/gapminder_all.csv')
````

This output is the rows in our CSV file, now as a pandas 'DataFrame' object. 

The first column is the **index** of the DataFrame. The index is used to identify the position of the data, but it is not an actual column of the DataFrame. 

It looks like the `read_csv` function in Pandas read our file properly. However, we havenâ€™t saved any data to memory so we can work with it. We need to assign the `DataFrame` to a variable. Remember that a variable is a name for a value, such as x, or data. We can create a new object with a variable name by assigning a value to it using the single equals sign (`=`).

````
data = pd.read_csv("data/gapminder_all.csv", index_col="country")
````

Above we supplied an optional **keyword argument** to the `read_csv` function: `index_col="country"`.

Keyword arguments look like variable assignments, but they happen inside the parentheses of a function call. They are especially useful when a function takes many optional arguments; they allow us to call the function with just the particular arguments we want. The names of the arguments (on the left side of the equals sign) are defined by the authors of the function we are using. 

When using `pandas`, it's good to familiarize yourself with [the documentation](https://pandas.pydata.org/docs/), which describes the required and optional arguments for each function in the library.

This particular argument to `read_csv`, `index_col`, indicates that we want to use the `country` column as the index of our `DataFrame`. Setting an index is not required but can be helpful. But be careful: you almost always want to select a column where the values are unique. So we would **not** want to set the `continent` column as the index, since those values repeat.

### DataFrame functions, columns, and types

There are many ways to summarize and access the data stored in DataFrames, using attributes and functions provided by the `DataFrame` object.

Functions are called on a DataFrame object using the syntax `df_object.function()`. As an example, `data.head()` gets the first few rows in the DataFrame referred to by the `data` variable, using the `head()` function.



````
data.head()
````

A `DataFrame` is a defined Python type. The parts of the type name preceding the word _DataFrame_ indicate that this is a custom (not built-in) type, defined as part of the `pandas` library.

````
type(data)
````

Like other Python collections, DataFrames can contain elements of diverse types. However, when all the elements of a single column have the same type, this column is considered to have the `dtype` of its members. 

If the column has mixed types of elements, or if all the elements are strings, the column has the generic `object` `dtype`.

````
data.dtypes
````

The types of a DataFrame matter because the `pandas` library is optimized for efficient numeric computation. As you can see from the above output, The `read_csv()` function actually converted the data in the numeric columns to values of the `float64` type (which is basically a high-powered version of `float`).

That's good news for us, because we don't have to bother converting them ourselves before doing calculations.

The `DataFrame.columns` variable stores information about the DataFrame's columns. 

Note that this is an attribute, not a function, so do not use `()` to try to call it. Attributes are like metadata about Python objects.

````
data.columns
````

The `DataFrame.describe()` function provides summary statistics for each numeric column in our dataset.

````
data.describe()
````

## Indexing and slicing a DataFrame

In the previous lesson, we worked with lists and dictionaries, and we saw how to build a nested data structure emulating our tabular data, so that we could access cell values by column name.

But such nested structures are still not optimal for working with data in table form. For instance, when we wanted to isolate the 2007 GDP per capita values for a subset of countries, we had to write a `for` loop and populate a new list.

DataFrames in `pandas` are designed to facilitate access to two-dimensional data. 

They also provide the best of both worlds from Python lists and dictionaries.

Thus, we can access parts of our DataFrame 
* by labels (column/index headings)
* by position (numeric values and ranges)
* by specific element locations (rows & columns)

### Selecting data by column
We use square brackets `[]` to select a single column by name. The object we get back is called a `Series`.

````
data['lifeExp_2007']
````

Note how the **index** is included in the `Series`. Because we set the index to the country names, we now have something very similar to the dictionary mapping country names to their population change that we created in yesterday's lesson.

We can also access the data in a column -- provided that the column name is also valid as a Python variable name (must begin with a letter, no special characters apart from the underscore, no spaces) -- as an attribute, using dot notation:

````
data.lifeExp_2007
````

### Selecting data by rows

Since we often want to filter our datasets to particular rows, `pandas` provides several ways of doing that.

The special object `.loc`, which is an attribute of every `DataFrame`, allows to us select a row by its index label. 

The syntax for `.loc` is a bit unusual: note that we place the label we want inside **square brackets**, not parentheses. You can think of `.loc` as a special kind of dictionary; the label is the key, and the value is the row indexed by that label.

````
data.loc['Algeria']
````

Note that a row is also represented as a `Series` object, except here the index labels are the **column** headings. 

We can achieve the same result by **position** using `.iloc`. Here, the `.iloc` object is acting sort of like a Python list.

````
data.iloc[0]
````

### Slicing a DataFrame

We can create slices of Python lists by using the syntax `[start:stop]`, where `start` is the index of the first element we want, and `stop` is the index of the next element after the last one that we want.

With a DataFrame, we can slice **both** by label **and** by position.

````
data.loc['Algeria':'Zimbabwe']
````

Note that this technique is only useful if your DataFrame is **already sorted** in some way. Our gapminder dataset is sorted by country name within each continent, so the above slice works to show us all the African countries.

We can also create slices with `.iloc`.

````
data.iloc[:5]
````

### Selecting values

To access a value at a specific row and column of a DataFrame, we have two options

Use `DataFrame.loc[indexLabel, columnName]` to select values by their labels.

````
data.loc['Angola', 'gdpPercap_2007']
````

Use `DataFrame.iloc[startIndex, stopIndex]` to select values by their numeric position. Positional indices start from 0, as in Python lists.

Here's the element in the second row, first column (not counting the index).


````
data.iloc[1, 0]
````

### Subsetting both rows and columns

To take a subset of both rows and columns at the same time, we can use the following syntax:
```
data.loc[firstLabel:lastLabel, [column1, column2, column3 ...]]
```
Where 
* `firstLabel` is the first index label we want
* `lastLabel` is the last index label we want
* `column1`, etc., are the names of the columns we want.

Our labels are strings, so we must put them in quotation marks.

**Note** that we include the names of the columns **inside a list**.

````
data.loc['Algeria':'Zimbabwe', ['lifeExp_1992','lifeExp_2002','lifeExp_2007']]
````

We can also supply a list of the index labels we want, and/or get a slice of the column labels.

The key in using `.loc` is the the **index labels must come first.**

````
data.loc[["Benin", "Turkey", "Afghanistan"],"lifeExp_1952":"lifeExp_2007"]
````

### Use Boolean expression to subset by criteria.

Slicing works well if the rows we want are contiguous. But what if we want to filter by some condition?

For instance, if we want to see only the data for countries where the 2007 life expectancy was greater than 80?

`pandas` has a special way of evaluating Boolean expressions. As a refresher from yesteray, a Boolean expression evaluates to either `True` or `False.


````
9.999 < 10
````

But look at what happens if we write a Boolean expression with a column from our DataFrame on the left-hand side.

````
data['lifeExp_2007'] > 80
````

Because we're selecting a column, we get back a `Series`, which includes the index of the original DataFrame. 

But instead of the numeric values in that column, we get Boolean values. The Series contains `True` where the values in the original column satisfy the Boolean expression, and `False` where they do not.

On its own, this isn't terribly useful. But in addition to passing labels to `data.loc`, we can also **pass a Boolean series aligned with the original DataFrame**. 

In other words, the `Series` we pass to `data.loc` has the same index as `data`. The result is a view of the original DataFrame where **only the rows marked `True` are shown**.

````
data.loc[data['lifeExp_2007'] > 80]
````

### Checking for null values

Some pandas functions also return a Boolean `Series`. An example is `data.isnull().any()`, which  returns `True` if any datum is missing and `False` otherwise.

The `any()` function accepts an optional keyword argument called `axis`. The **axis** of a DataFrame refers to whether the evaluation is done by rows or columns.

Typically, the default value for the argument -- meaning what it is assigned by the function if you don't include it is `axis=0`, which evaluates by rows.

````
data.isnull().any()
````

The above produces a `True`/`False` value for each column, according to whether any of the rows have null values in that column.

Often, we want to see any **rows** that have null values. To do that, we can use `data.isnull.any(axis=1)` and pass the result to `data.loc`.

````
null_rows = data.isnull().any(axis=1)
data.loc[null_rows]
````

In this case, the result is an **empty** DataFrame, indicating that **no rows** matched the null condition. This is a very clean dataset!

## Subsetting and aggregating data

Letâ€™s say we want to have a clearer view of how the European countries are split according to their GDP per capita.

We can split the countries in two groups across the years surveyed, those who presented a GDP higher than the European average and those with a lower GDP.

In addition to `.loc`, Our DataFrame has some built-in functions that facilitate grouping and calculating aggregate results (like averages).

If we were to do this with the tools we learned yesterday, we would need to rely on `for` loops. One of the advantages of the `pandas` library is that it reduces the need to use Python loops for grouping and aggregating data. Instead, `pandas` uses highly optimized code (written in C) to perform repetitive operations.

This additional layer of abstraction not only reduces the amount of code we have to write, but with large datasets, it speeds up computation (sometimes quite significantly).



### Creating a subset

Here we create a subset of the gapminder data that corresponds to the European countries. 

The only new function here is `copy()`. This is important because `.loc` normally returns a **view** of the original data (to avoid computational overhead). It's like when you filter a spreadsheet in Excel; the other rows are still there, just hidden. 

This behavior of `.loc` is different from how Python ordinarily handles slicing: slicing a Python list actually creates a copy of the original list.

With a DataFrame, we have to state expicitly that we want a copy.

````
europe = data.loc[data.continent == 'Europe'].copy()
````

Since we are working with the GDP per capita, we can isolate those columns, too.

The intial `:` in our `.loc` expression indicates that we want all the rows from our `europe` DataFrame.



````
europe = europe.loc[:, 'gdpPercap_1952':'gdpPercap_2007'].copy()
````

````
europe
````

### Aggregating by column

DataFrames come with several functions for aggregating values. Now that our `europe` DataFrame has just the rows and columns we want, we can use these functions to summarize our datset.

First, we'll find the **mean GDP per capita by year** (across all European countries).

We can write `europe.mean()` to compute the average value for each column. (Look, no `for` loops!)


````
mean_gdp = europe.mean()
print(mean_gdp)
````

(The default behavior of these functions, like `data.isnull().any()` above, applies the calculation to all the rows by column. To compute the values for all columns by row, use the `axis=1` keyword argument.)


````
mean_country = europe.mean(axis=1)
#mean_country2 = europe.mean(axis="columns")

print(mean_country)
````

Let's get some more practice with subsetting and aggregating. We want to know what the maximum life expectancy was in each year, and which country is represented by that figure. 

We have to re-create a dataframe with the countries in Europe because we overwrote the previous dataframe with our slice to get only the gdp columns. 

We'll first create a subset of the original data, with just the rows where the continent is "Europe". Then we'll use .loc on that to further extract the life expectancy columns.  You can chain together filters on the dataframe in this way. 

````
europe_life = data[data.continent == 'Europe'].loc[:, 'lifeExp_1952':'lifeExp_2007'].copy()
europe_life
````

Now we can use pandas aggregations to perform some calculations. We can use the `max()` method on a dataframe to get the highest values along an axis. 

By not passing an argument to `max()` we're using the default of axis=0 and getting the max in each column. 

````
lifeExp_max = europe_life.max()
print(lifeExp_max)

#This is a Series rather than a DataFrame. The index of the Series is the column label. 
````

What if instead of getting the value, we want the country which had the highest value in each column?  There is an aggregation called `idxmax()` which returns the index instead of the value. In our DataFrame, the country name is the index. 

````
lifeExp_max_country = europe_life.idxmax()
print(lifeExp_max_country)

#This is also a Series rather than a DataFrame. The index of the Series is the column label.
````

We have two series, but what if we wanted to see both the value and the index name? 

We can merge these two series into a DataFrame. Merge on axis=1 because we want to add these as columns. 

````
lifeexp_df = pd.concat([lifeExp_max, lifeExp_max_country], axis=1)
lifeexp_df
````

## Grouping and plotting

Frequently, the data we're using can be subsetted into discrete groups. Above, we saw how to isolate just the countries in Europe using `.loc`.

What if we wanted to look at GDP values for each continent?

It's called `groupby`.

Before grouping, however, let's isolate the GDP columns, as we did above. (It's easier to slice your data **before** grouping than after.)

We're including the `continent` column in our slice, because we want to be able to group on it.

````
gdp_data = data.loc[:, 'continent':'gdpPercap_2007']
grouped = gdp_data.groupby('continent')
grouped
````

Here we called the `groupby` function . The `groupby` function takes as an argument the name of one or more columns.

These columns will almost always have **categorical** values -- meaning, discrete values that repeat. In this case, the `continent` column contains the names of the continents, which are discrete and do repeat. `groupby` will use these values to create groups out of the original dataset. 

Note that calling `groupby` doesn't actually produce an informative result. That's because we need to tell `pandas` what to do with the groups.

Usually, we'll want to apply an aggregate function -- like `mean()` -- either to a single column or to all the columns from the original dataset.

The result will show the aggregate result of that operation **for each group**.



````
gdp_means = grouped.mean()
````

````
gdp_means
````

### Cleaning up column names

In order to plot the mean GDP per capita by continent over time, it would be convenient if our column names just indicated the years. Otherwise, our plot will not look very good.

In order to remove the prefix from the column names, we can use the Python string function `strip`.

With a DataFrame we can manipulate the columns directly by using the `columns` attribute. 

`gdp_means.columns` returns a special pandas `Index` object. We could use a `for` loop to `strip` each column header in sequence. 

But there's a faster way. We can call the `strip` function directly on the `columns` object by way of its `str` attribute. (This works only when all the values in the sequence are Python strings.)



````
columns = gdp_means.columns.str.strip('gdpPercap_')
````

While we're at it, let's convert the column headers from string to integer values. Since the headers represent years, we can use them as a time series in our plot -- along the x axis. 

Converting them to integers will ensure that they are ordered correctly on the axis.

We've used the `int` function before to convert a string to an interger. With `pandas` objects, we can use the `astype` function to do the same thing. `astype` takes as its argument the name of Python type (in quotation marks).

````
columns = columns.astype('int')
````

Now we just need to re-assign our cleaned-up columns to the `gdp_means` DataFrame.

````
gdp_means.columns = columns
````

````
gdp_means
````

### Plotting with `pandas` and `matplotlib`

A DataFrame has a built-in `plot` function, which uses the `matplotlib` library under the hood to create basic charts and graphs.

The `plot` function defaults to a line graph, with the DataFrame's index as the x axis. In this case, we want the **columns** to be the x axis. 

Fortunately, we can **transpose** our DataFrame using the `.T` (capital "T") attribute defined on every DataFrame. Transposing a DataFrame yields a new DataFrame, where the index labels are the column headings, and the column headings are the index labels.

````
from matplotlib import pyplot as plt
````

````
gdp_means.T.plot()
````

`matplotlib` allows a fair amount of flexibility in customizing your plot. To set custom attributes, we can use the `AxesSubplot` object, which is returned by the call `DataFrame.plot`.

````
ax = gdp_means.T.plot()
ax.set_title('GDP Per Capita by Continent')
ax.set_xlabel('Year')
ax.set_ylabel('Dollars');
````

We're not using the built-in `plot` method of our DataFrame anymore, so we need explicitly to tell seaborn's `scatterplot` method which data to use.

By referencing the columns of our DataFrame `data`, we can assign them to the axes of our scatterplot.

To start with, let's plot GDP per capita against life expectancy for 2007.

````
ax = sns.scatterplot(data=data,
                x=data.gdpPercap_2007,
                y=data.lifeExp_2007)
````

We can begin to see an interesting distribution curve between GDP per capita and life expectancy. But since all the data points are the same size and color, it's hard to determine what the individual dots might represent, vis-a-vis country.

Our points are differentiated by size, but they're still relatively small. We can adjust this by providing a pair of numeric values to the optional `sizes` keyword argument.

We enclose these values in parentheses: this is a Python data type called a `tuple`.

And finally, let's give our plot a title.

````
ax = sns.scatterplot(data=data,
                x=data.gdpPercap_2007,
                y=data.lifeExp_2007,
                hue=data.continent,
                size=data.pop_2007/1e6,
                alpha=0.75,
                sizes=(100,1000))
ax.legend(labelspacing=2)
ax.set_title('GDP per capita vs. life expectancy by country.')
````

If our plot is too small, we can fix this by setting the `figure.figsize` attribute of `plt.rcParams`, where `plt` is a reference to the imported `pyplot` library.

### More fun with Boolean expressions

We define the **wealth score** as the average measure for a given country of how often its GDP per capita was above the European mean.

So a country with wealth score of 1 would be a country whose GDP per capita was greater than the continental mean for every year in our dataset. 

In order to calculate the wealth score for each country, we first need to determine whether a given country's per capita GDP for a given year was above or below that year's average (mean).

Above, we saw what happens when we write a Boolean expression to compare a column of a DataFrame (which is a pandas `Series`) to a single value.

The code 
```
data.gdpPercap_2007 <= 1000
```
yields a Series whose values are either `True` or `False`, depending on whether each element in the `gdpPercap_2007` column is less than 1000.

We can also compare a Series and a DataFrame. The following code compares **each row** of our DataFrame to the **Series of mean values from each column**.

The column labels from the DataFrame `europe` will be lined up with the index labels in `mean_gdp`.  

### Boolean addition

`gdp_comp` is a new DataFrame. Note that it has the same dimensions as the DataFrame from which was derived (`europe`). 

The values in each cell indicate whether the value in that cell is greater than the mean value for that cell's column. 

How do we turn this information into a single numeric score for each country?

We can exploit a basic feature of Python (and other programming languages): `True` and `False` are binary values, meaning they are functionally equivalent to 1 and 0.

The following code looks odd, but it works, because in a numeric context, Python Booleans evaluate to their numeric equivalents.

(The built-in `sum` function sums the numbers in a list.)

````
gdp_comp = europe > mean_gdp
gdp_comp
````

````
sum([True, False, True, True])
````

A DataFrame also has a `sum` function, which by default sums the values in each column. But in this case, we want to evaluate the wealth score for each country (in other words, for each row).

We can use the `axis=` keyword argument to produce a new Series representing the sum of the Boolean values in each row of our `gdp_comp` DataFrame.

````
gdp_comp.sum(axis=1)
````

The numbers here tell us how many `True` values were in each row (since `False` values evaluate to 0). 

And the number of `True` values tells us, in turn, how many times the GDP per capita (the values in the original columns of our `europe` DataFrame) exceeded the mean for each column.

Now all we need to do is average these numbers. 

````
wealth_scores = gdp_comp.sum(axis=1) / len(mean_gdp)
wealth_scores
````

### DataFrame calculations

Much as we can write Boolean expressions with Series and DataFrames using the standard Python comparison operators (`==`, `<`, `>`, etc.), we can do math with DataFrames and Series using the standard Python numeric operators for addition, subtraction, etc.

Dividing a Series by a single value will compute the division for every element in the Series, creating a new Series out of the results.

Here we use `len` (the built-in function) to get the number of mean values (or the number of GDP-per-capita columns) against which we compared the data for each country.

Our new Series `wealth_scores` has the same index as our `europe` DataFrame. So we can add the Series to the latter as a new column.

````
europe['wealth_score'] = wealth_scores
````
